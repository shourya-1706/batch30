```python
people = ['a','b','c','d','e','f','g','h']
age = [21,17,12,33,55,40,15,18]
weight = [55,35,57,11,10,16,70,82]
height = [160,171,180,182,165,173,156,154]



import matplotlib.pyplot as plt



plt.scatter(age , height)
plt.title("relationship between age and height of patients")
plt.ylabel("height in masters")
plt.xlabel("age of patients")
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.title("people's weight in kgs",fontsize=16,fontstyle='italic')


# main plot function 'bar'
plt.bar(x=people,height=weight,width=0.6,color='black',edgecolor='k',alpha=0.6)

plt.xlabel("people",fontsize=15)
plt.xticks(fontsize=15,rotation=15)
plt.yticks(fontsize=14)
plt.ylabel("weight(in kgs)",fontsize=15)
```
